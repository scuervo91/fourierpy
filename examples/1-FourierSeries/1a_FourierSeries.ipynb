# Fourier Series Calculations

```python
import numpy as np
from fourierpy import FourierModel
import matplotlib.pyplot as plt
```

```python
## Aprroximate functions
```

```python
def square(x):
    return np.square(x)
```

```python
time = np.linspace(0,2*np.pi,1000)
y = square(time)

plt.plot(time,y)
```

```python
f1 = FourierModel()
f1.fit(time,y,5)
f1
```

```python
time_fs1 =np.linspace(0,2*np.pi,1000)

fs1 = f1.get_series(time_fs1)
plt.plot(time,y, label ='original')
plt.plot(time_fs1,fs1, label ='Fourier')
plt.legend()
```

```python
## Discontinious functions
```

```python
x2 = np.linspace(0,20,1000)
y2 = np.zeros(1000)
y2[x2>15] = 1
y2[x2<12] = 7.45

f2 = FourierModel()
f2.fit(x2,y2,8)

fs2 = f2.get_series(x2)

plt.plot(x2,y2, label ='original')
plt.plot(x2,fs2, label ='Fourier')
plt.legend()
```

```python
x3 = np.linspace(0,20,1000)
y3 = np.zeros(1000)
y3[x3>15] = 1
y3[x3>16] = 0

f3 = FourierModel()
f3.fit(x3,y3,8)

fs3 = f3.get_series(x3)

plt.plot(x3,y3, label ='original')
plt.plot(x3,fs3, label ='Fourier')
plt.legend()
```
